```python
# One-hot encode the categorical columns
categorical_cols = ['Fuel_Type', 'Seller_Type', 'Transmission','Year','Owner']
onehotencoder = OneHotEncoder(sparse_output=False, drop='first')  # Use drop='first' to avoid dummy variable trap
encoded_categorical = onehotencoder.fit_transform(dataset[categorical_cols])

# Get the names of the one-hot encoded columns
encoded_categorical_cols = onehotencoder.get_feature_names_out(categorical_cols)

# Convert the encoded array to a DataFrame
encoded_categorical_df = pd.DataFrame(encoded_categorical, columns=encoded_categorical_cols)

# Combine the encoded categorical columns with the numeric columns
final_dataset = pd.concat([encoded_categorical_df, dataset[numeric_cols].reset_index(drop=True)], axis=1)
```

```python
x_train,x_test,y_train,y_test = train_test_split(final_dataset,dataset['Selling_Price'],test_size=0.2,random_state=0)
```

```python
import torch
x_train_torch = torch.tensor(x_train.values, dtype=torch.float32)
y_train_torch = torch.tensor(y_train.values, dtype=torch.float32)

x_test_torch = torch.tensor(x_test.values, dtype=torch.float32)
y_test_torch = torch.tensor(y_test.values, dtype=torch.float32)
```

```python
from torch.utils.data import Dataset,DataLoader
class CustomDataset(Dataset):
    def __init__(self,x,y):
        self.x = x
        self.y = y
    def __getitem__(self,index):
        return self.x[index],self.y[index]
    def __len__(self):
        return len(self.x)
    
x_train_dataset = CustomDataset(x_train_torch,y_train_torch)
x_test_dataset = CustomDataset(x_test_torch,y_test_torch)

x_train_dataloader = DataLoader(x_train_dataset,batch_size=32,shuffle=True)
x_test_dataloader = DataLoader(x_test_dataset,batch_size=32,shuffle=True)
```

```python
for x,y in x_train_dataloader: print(x.shape); break #x_train_dataloader
```

```python
class SimpleANN(torch.nn.Module):
    def __init__(self):
        super(SimpleANN, self).__init__()
        self.fc1 = torch.nn.Linear(23, 6)
        self.relu = torch.nn.ReLU()
        self.fc2 = torch.nn.Linear(6, 1)
    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x
    
model = SimpleANN()

criterion = torch.nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(100):
    for x, y in x_train_dataloader:
        y_pred = model(x)
        optimizer.zero_grad()
        loss = criterion(y_pred, y.unsqueeze(1))
        loss.backward()
        optimizer.step()
    print(f'Epoch {epoch+1}, Loss: {loss.item()}')

with torch.no_grad():
    for x, y in x_test_dataloader:
        y_pred = model(x)
        loss = criterion(y_pred, y.unsqueeze(1))
        print(f'Test Loss: {loss.item()}')
```
