```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder
from sklearn import datasets
```

```python
data = datasets.load_breast_cancer()
```

```python
x,y = data.data, data.target
dfdata = pd.DataFrame(x)
dflabels = pd.DataFrame(y)

X_train, X_test, y_train, y_test = train_test_split(dfdata, dflabels, test_size=0.2)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
lableencoder = LabelEncoder()
y_train = lableencoder.fit_transform(y_train)
y_test = lableencoder.transform(y_test)
```

```python
import torch 
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.utils.data import DataLoader
from torch.utils.data import TensorDataset
```

```python
x_train_tensor = torch.tensor(X_train,dtype=torch.float32)
y_train_tensor = torch.tensor(y_train,dtype=torch.float32)
x_test_tensor = torch.tensor(X_test,dtype=torch.float32)
y_test_tensor = torch.tensor(y_test,dtype=torch.float32)
```

```python
class SimpleNN(nn.Module):
    def __init__(self,x):
        self.weights = torch.rand(x.shape[1],1,dtype=torch.float32,requires_grad=True)
        self.bias = torch.ones(1,dtype=torch.float32,requires_grad=True)

    def forward(self,x):
        # print(f"shape of x is {x.shape}")
        z=  torch.matmul(x,self.weights) + self.bias
        # print(f"shape of z is {z.shape} {z}")
        return torch.sigmoid(z)
    def loss(self,y_pred,y):
        return F.binary_cross_entropy(y_pred,y)
```

```python
simple = SimpleNN(x_train_tensor)
for i in range(1):
    result=simple.forward(x_train_tensor)
    loss = simple.loss(result, y_train_tensor.reshape(-1,1))
    print(f"{loss}")
    loss.backward()
    with torch.no_grad():
        simple.weights -= 0.1 * simple.weights.grad
        simple.bias -= 0.1 * simple.bias.grad
    simple.weights.grad.zero_()
    simple.bias.grad.zero_()
    print(f"{result} {y_train_tensor}")
```
